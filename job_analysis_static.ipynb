# Job Market Analysis

## Overview
This project analyzes job postings collected from TimesJobs to identify key trends in the job market.  
The analysis provides insights into the top hiring companies, most in-demand skills, popular job locations, and job roles.

## Dataset
The dataset contains information such as:
- **Company Name:** The name of the company posting the job.
- **Skills:** Required skills for the job.
- **Location:** Where the job is located.
- **Job Type:** The category or role of the job.
- **Job Link:** A direct link to the job listing.
- **Posted Date:** When the job was posted.


## Goals of the Analysis
- Identify the top companies hiring.
- Analyze the most in-demand skills.
- Discover the most popular job locations.
- Understand the top job roles available.



## Libraries Used
The following Python libraries were used in the analysis:
- `pandas` for data manipulation.
- `matplotlib` and `seaborn` for data visualization.


 ---

### Import Necessary Libraries
First, we import the required libraries for data manipulation and visualization.


```python
# Import necessary libraries
import pandas as pd  # For data manipulation
import matplotlib.pyplot as plt  # For plotting
import seaborn as sns  # For enhanced visualizations

# Set a dark background theme for better visualization
plt.style.use('dark_background')

# Load the job data from CSV file
job_df = pd.read_csv('data/job_data.csv')

# Display the first few rows to check the loaded data (optional)
# job_df.head()
```



### Loading the Processed Data
The dataset used in this analysis has been preprocessed in a separate file to handle missing values, inconsistencies, and formatting issues. In this notebook, we will focus on exploring and visualizing the cleaned data.

The dataset is loaded from the `data/job_data.csv` file for further analysis.

---


### Top 20 Companies with Most Job Postings

In this analysis, we identify the top 20 companies with the highest number of job postings. 

- We group the job data by `Company Name` and count the number of job postings for each company.
- The data is sorted in descending order to highlight the top hiring companies.
- A horizontal bar chart is used for better visualization and comparison.
- The chart provides insights into which companies are actively hiring the most.





```python
# Grouping the job data by 'Company Name' and counting the number of job postings
most_jobs_cmp = job_df.groupby('Company Name')['Job Link'].count().sort_values(ascending=False).head(20)

# Resetting the index for better readability
most_jobs_cmp = most_jobs_cmp.reset_index()

# Renaming columns for better understanding
most_jobs_cmp.columns = ['Company Name', 'Number of Jobs']

# Plotting a horizontal bar chart to visualize the top 20 hiring companies
most_jobs_cmp.plot(kind='barh', x='Company Name', y='Number of Jobs', figsize=(15, 6), color='#FFA53F')

# Adding labels and title for clarity
plt.ylabel('Number of Jobs', fontsize=12)
plt.xlabel('Company Name', fontsize=12)
plt.grid(linestyle='--', alpha=0.6)
plt.title('Top 20 Companies with Most Jobs', fontsize=15, fontweight='bold')

# Inverting the y-axis to show the highest values on top
plt.gca().invert_yaxis()

# Displaying the plot
plt.show()
```

**Key Insights:**
- The companies listed here have the highest number of job postings in the dataset.
- This information can help job seekers target companies with high hiring potential.
- Industries with the most active hiring can also be inferred from this visualization.

---


### Top 20 In-Demand Skills

This visualization showcases the most sought-after skills based on the job postings in our dataset.  
Understanding the top required skills helps job seekers align their skill sets with market demands.






```python
# Splitting the 'Skills' column to extract individual skills and counting their occurrences
skills = job_df['Skills'].str.split(', ', expand=True).stack().value_counts().head(20)

# Resetting the index for better visualization
skills = skills.reset_index()

# Renaming the columns for better readability
skills.columns = ['Skill', 'Count']

# Plotting a horizontal bar chart to visualize the top 20 required skills
skills.plot(kind='barh', x='Skill', y='Count', figsize=(15, 6), color='lightblue')

# Adding labels and title for better understanding
plt.ylabel('Skill', fontsize=12)
plt.xlabel('Count', fontsize=12)
plt.grid(linestyle='--', alpha=0.6)
plt.title('Top 20 Skills Required', fontsize=15, fontweight='bold')

# Inverting the y-axis to display the most required skills at the top
plt.gca().invert_yaxis()

# Displaying the plot
plt.show()
```

**Key Insights:**
- The top 20 skills are highlighted, giving a quick view of what's currently in demand.
- Job seekers can use this data to focus on developing skills that align with job market trends.
- The demand for specific technical and soft skills can be observed.

### Top 20 Job Hotspots

This visualization displays the top 20 locations with the highest number of job postings.  
Understanding where jobs are concentrated can help job seekers target the most promising areas.

```python
# Counting job postings by location and selecting the top 20 locations
hotspots = job_df['Location'].value_counts().head(20)

# Resetting the index for better visualization
hotspots = hotspots.reset_index()

# Renaming columns for better readability
hotspots.columns = ['Location', 'Count']

# Removing any empty location entries
hotspots = hotspots[hotspots['Location'] != ""]

# Plotting a horizontal bar chart to visualize job distribution by location
hotspots.plot(kind='barh', x='Location', y='Count', figsize=(15, 6), color='lightgreen')

# Adding labels and title for better understanding
plt.ylabel('Location', fontsize=12)
plt.xlabel('Count', fontsize=12)
plt.grid(linestyle='--', alpha=0.6)
plt.title('Top 20 Job Hotspots', fontsize=15, fontweight='bold')

# Inverting the y-axis to display the locations with the highest job postings at the top
plt.gca().invert_yaxis()

# Displaying the plot
plt.show()
```

**Key Insights:**
- The data reveals which cities or regions have the most job opportunities.
- This insight can help individuals plan job searches based on high-demand locations.
- Businesses can also leverage this data to focus recruitment efforts in key regions.

### Top 20 Job Roles

This visualization displays the top 20 most common job roles found in the dataset.  
Understanding job role distribution helps professionals align their career paths with market demand.

```python
# Counting occurrences of each job type and selecting the top 20
top_roles = job_df['Job Type'].value_counts().head(20)

# Resetting the index for better visualization
top_roles = top_roles.reset_index()

# Renaming columns for clarity
top_roles.columns = ['Job Type', 'Count']

# Plotting a horizontal bar chart to visualize the most common job roles
top_roles.plot(kind='barh', x='Job Type', y='Count', figsize=(15, 6), color='lightcoral')

# Adding labels and title for better understanding
plt.ylabel('Job Type', fontsize=12)
plt.xlabel('Count', fontsize=12)
plt.grid(linestyle='--', alpha=0.6)
plt.title('Top 20 Job Roles', fontsize=15, fontweight='bold')

# Inverting the y-axis to display the top job roles at the top
plt.gca().invert_yaxis()

# Displaying the plot
plt.show()
```

**Key Insights:**
- The chart helps identify the most sought-after job roles in the market.
- It can guide job seekers in targeting roles that are in high demand.
- Employers can use this data to benchmark their hiring strategies.
